```python
#Khai báo thư viện

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Vẽ các đồ thị bên trong cell

%matplotlib inline
```

```python
#Lấy dataset

tips_df = sns.load_dataset('tips')
tips_df.head()
```

# 4. Thiết kế biểu đồ với Seaborn

## 4.1. Thiết lập theme

```python
# Cú pháp:
#     sns.set_theme() => Mặc định
#     sns.set_style("darkgrid/whitegrid/dark/white/ticks") => Tùy chỉnh

# BEFORE
sns.kdeplot(data = tips_df, x ='total_bill', hue = 'sex',
            multiple = 'layer').set(title = 'TOTAL BILL COVERED BY GENDER')
```

```python
# Cú pháp:
#     sns.set_theme() => Mặc định
#     sns.set_style("darkgrid/whitegrid/dark/white/ticks") => Tùy chỉnh

# AFTER
sns.kdeplot(data = tips_df, x ='total_bill', hue = 'sex',
            multiple = 'stack').set(title = 'TOTAL BILL COVERED BY GENDER')

sns.set_style('dark')
```

## 4.2. Thiết lập palette

```python
# Cú pháp:
#     palette = 'reds/purples...'
#     palette = 'tab10/set1,2,3'
#     palette = 'viridis...'

# BEFORE
sns.kdeplot(data = tips_df, x ='total_bill', hue = 'sex', multiple = 'stack').set(title = 'TOTAL BILL COVERED BY GENDER')
```

```python
# Cú pháp:
#     palette = 'reds/purples...'
#     palette = 'tab10/set1,2,3'
#     palette = 'viridis...'

# AFTER
sns.kdeplot(data = tips_df, x ='total_bill', hue = 'sex', multiple = 'stack',
            palette = 'viridis').set(title = 'TOTAL BILL COVERED BY GENDER')
```

## 4.3. Thiết lập size

### 4.3.1. Hisplot

```python
# Cú pháp: binwidth = kích thước

# BEFORE
sns.histplot(data = tips_df, x ='total_bill', hue = 'sex', multiple = 'stack').set(title = 'TOTAL BILL COVERED BY GENDER')
```

```python
# Cú pháp: binwidth = kích thước

# AFTER
sns.histplot(data = tips_df, x ='total_bill', hue = 'sex', multiple = 'stack',
             binwidth = 7, palette = 'viridis').set(title = 'TOTAL BILL COVERED BY GENDER')
```

### 4.3.2. Distplot

```python
# Cú pháp: height = kích thước, aspect = kích thước

# BEFORE
tbill = sns.displot(data = tips_df, x = 'total_bill', col = 'time', hue = 'sex', multiple='stack', kde = True)

tbill.fig.suptitle('TOTAL BILL COVERED BY GENDER')
tbill.fig.subplots_adjust(top = .9)
```

```python
# Cú pháp: height = kích thước, aspect = kích thước

# AFTER
tbill= sns.displot(data = tips_df, x = 'total_bill', col = 'time', hue = 'sex', multiple='stack', kde = True,
                   height = 6, aspect = 1)

tbill.fig.suptitle('TOTAL BILL COVERED BY GENDER')
tbill.fig.subplots_adjust(top = .9)
```

### 4.3.3. FacetGrid

```python
# Cú pháp: height = kích thước, aspect = kích thước, s = kích thước

# BEFORE
fg = sns.FacetGrid(data = tips_df, hue = 'smoker')
fg.map_dataframe(sns.scatterplot, x = "total_bill", y = "tip").set(title = 'TIP AND TOTAL BILL COVERED BY SMOKER')
fg.add_legend() 
```

```python
# Cú pháp: height = kích thước, aspect = kích thước, s = kích thước

# AFTER
fg = sns.FacetGrid(data=tips_df, hue = 'smoker', height = 5, aspect = 1.4)
fg.map_dataframe(sns.scatterplot, x = "total_bill", y = "tip", s = 80).set(title = 'TIP AND TOTAL BILL COVERED BY SMOKER')
fg.add_legend() 
```

### 4.3.4. Jointplot

```python
# Cú pháp: height = kích thước, ratio = kích thước

# BEFORE
tb = sns.jointplot(data = tips_df, x = 'tip', y = 'total_bill', hue = 'smoker', ylim = [0,60])
tb.fig.suptitle('CORRELATION BETWEEN TOTAL BILL AND TIP BY SMOKER') 
tb.fig.subplots_adjust(top = .9) 
```

```python
# Cú pháp: height = kích thước, ratio = kích thước

# AFTER
tb = sns.jointplot(data = tips_df, x = 'tip', y = 'total_bill', hue = 'smoker', ylim = [0,60],
                  height = 8, ratio = 4)
tb.fig.suptitle('CORRELATION BETWEEN TOTAL BILL AND TIP BY SMOKER') 
tb.fig.subplots_adjust(top = .9) 
```

```python
pg_df = sns.load_dataset('penguins')
pg_df.tail()
```

### 4.3.5. LM Plot

```python
# Cú pháp: height = kích thước, aspect = kích thước 

# BEFORE
tb = sns.lmplot(data = tips_df, x = 'total_bill', y = 'tip', hue = 'smoker')
tb.fig.suptitle('CORRELATION BETWEEN TOTAL BILL  AND TIP BY SMOKER') 
tb.fig.subplots_adjust(top = .9) 
```

```python
# Cú pháp: height = kích thước, aspect = kích thước 

# AFTER
tb = sns.lmplot(data = tips_df, x = 'total_bill', y = 'tip', hue = 'smoker',
                height = 6, aspect = 1.2)

tb.fig.suptitle('CORRELATION BETWEEN TOTAL BILL  AND TIP BY SMOKER') 
tb.fig.subplots_adjust(top = .9) 
```

### 4.3.5. Pairplot

```python
# Cú pháp: height = kích thước, aspect = kích thước

# BEFORE
pg = sns.pairplot(data = pg_df, hue = 'species', markers = ['o','s','D'])
pg.fig.suptitle('CORRELATION BETWEEN BODY PARTS OF 3 PENGUINS') 
pg.fig.subplots_adjust(top = .9) 
```

```python
# Cú pháp: height = kích thước, aspect = kích thước

# AFTER
pq = sns.pairplot(data = pg_df, hue = 'species', markers = ['o','s','D'],
                 height = 3, aspect = 1)
pq.fig.suptitle('CORRELATION BETWEEN BODY PARTS OF 3 PENGUINS') 
pq.fig.subplots_adjust(top = .9) 
```

```python
fl_df = sns.load_dataset('flights')
fl_df.head()
```

```python
fls = pd.pivot_table(data = fl_df, index = 'month', columns = 'year', values = 'passengers')
fls.head()
```

### 4.3.6. Heatmap

```python
# Cú pháp:

#     linewidths = kích thước
#     fig,ax = plt.subplots(figsize = (kích thước 1, kích thước 2))
#     annot_kws = {'fontsize': kích thước}

#     title = 'TÊN BIỂU ĐỒ'
#     ax.set_title(title, fontsize = kích thước)

# BEFORE
sns.heatmap(data = fls, annot = True, fmt = 'd', vmin = 0, vmax = 700).set(title = 'PASSENGERS OF FLIGHTS FROM 1949 TO 1960')
```

```python
# Cú pháp: 

#     linewidths = kích thước
#     fig,ax = plt.subplots(figsize = (kích thước 1, kích thước 2))
#     annot_kws = {'fontsize': kích thước}

#     title = 'TÊN BIỂU ĐỒ'
#     ax.set_title(title, fontsize = kích thước)

# AFTER
fig,ax = plt.subplots(figsize = (15,10))
sns.heatmap(data = fls, annot = True, fmt = 'd',
            vmin = 0, vmax = 700,
            linewidths = 3, annot_kws = {'fontsize':16})

title = 'PASSENGERS OF FLIGHTS FROM 1949 TO 1960'
ax.set_title(title, fontsize = 18)
```
