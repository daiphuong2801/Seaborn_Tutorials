### 3.2.1. Distribution Plot

### a. Histogram Plot

```python
#Cú pháp: sns.hisplot(data = Tên cần đặt_df, x = 'Chỉ tiêu')

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
# multiple (layer, dodge, stack, fill):
# stat (count, frequency, probability/proportion, percent, density):
# kde = True/False: 
```

```python
sns.histplot(data=tips_df, x ="total_bill", multiple="dodge",stat = 'percent', kde = False, hue = 'sex')
```

### b. KDE (Kernel Density Estimate) Plot

```python
#Cú pháp: sns.kdeplot(data=Tên cần đặt_df, x = 'Chỉ tiêu')

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
# multiple (layer, stack, fill)
# stat (density) => mặc định
```

```python
sns.kdeplot(data=tips_df, x ='total_bill', hue = 'sex', multiple = 'layer')
```

### c. Displot

```python
#Cú pháp: sns.displot(data=Tên cần đặt_df, x = 'Chỉ tiêu')

#Các parameter bổ sung:
# col: phân biệt chỉ tiêu bất kỳ theo biểu đồ 
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
# multiple (layer, dodge, stack, fill):
# kde = True/False:
# stat (count) => mặc định
```

```python
sns.displot(data=tips_df, x = 'total_bill',col = 'time', hue = 'sex', multiple='stack', kde = True)
```

### 3.2.2. Bar Plot

```python
#Cú pháp: sns.barplot(data=Tên cần đặt_df, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2')

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
```

```python
sns.barplot(data=tips_df, x = 'day', y = 'total_bill', hue = 'sex', hue_order = ['Male', 'Female'])
```

### 3.2.4. Box Plot

```python
#Cú pháp: sns.boxplot(data=Tên cần đặt_df, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2')

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
```

```python
sns.boxplot(data = tips_df, x = 'day', y = 'total_bill', hue = 'smoker')
```

### 3.2.5. FacetGrid

```python
#Cú pháp: 
# Tên class = sns.FacetGrid(data = Tên cần đặt_df)
# Tên class.map_dataframe(sns.histplot/scatterplot, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2' )

#Các parameter bổ sung:
# col: phân biệt chỉ tiêu bất kỳ theo cột
# row: phân biệt chỉ tiêu bất kỳ theo dòng
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue/row/col_order: hoán đổi thứ tự của hue/row/col
```

```python
#Scatter Plot

fg = sns.FacetGrid(data=tips_df, hue = 'smoker')
fg.map_dataframe(sns.scatterplot, x = "total_bill", y = "tip")
fg.add_legend() # hiển thị chú thích khi có hue
```

```python
#Histogram Plot

fg = sns.FacetGrid(data = tips_df, col = "time",  row = "sex")
fg.map_dataframe(sns.histplot, x = "total_bill")
```

### 3.2.6. Joint Plot

```python
pg_df = sns.load_dataset('penguins')
pg_df.tail()
```

```python
#Cú pháp: sns.jointplot(data=Tên cần đặt_df, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2')

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# hue_order: hoán đổi thứ tự của hue
# kind (scatter, kde, hist, hex, reg, resid): kiểu biểu đồ muốn hiển thị (hue không đi kèm kind = 'hex', 'reg', 'resid')
```

```python
sns.jointplot(data=pg_df, x="flipper_length_mm", y="bill_length_mm", hue='species' , kind = 'kde')
```

### 3.2.7. Pair Plot

```python
#Cú pháp: sns.pairplot(data = Tên cần đặt_df) 

#Các parameter bổ sung:
# hue: phân biệt chỉ tiêu bất kỳ theo màu
# kind (scatter, kde, hist, reg): kiểu biểu đồ muốn hiển thị => mặc định: kind = 'scatter'
# diag_kind (auto, hist, kde, None): kiểu biểu đồ muốn hiển thị đối với đường chéo chính
# markers (o,s,D): thay đổi ký hiệu chú thích (của hue)
```

```python
sns.pairplot(data = pg_df, hue = 'species', diag_kind = 'hist', markers = ['o','s','D'])
```

```python
sns.pairplot(data = pg_df, hue = 'species', kind = 'kde')
```

### 3.2.8. Heat Map

```python
fl_df = sns.load_dataset('flights')
fl_df
```

```python
#Sắp xếp dữ liệu 

#Cú pháp: Tên table = pd.pivot_table(data = Tên cần đặt_df, index = 'Chỉ tiêu theo hàng', columns = 'Chỉ tiêu theo cột', values = 'Giá trị')
```

```python
fls = pd.pivot_table(data = fl_df, index = 'month', columns = 'year', values = 'passengers')
fls
```

```python
#Cú pháp: sns.heatmap(data = Tên cần đặt) 

#Các parameter bổ sung:
# annot: hiển thị từng giá trị vào biểu đồ
# fmt (d,f): thay đổi kiểu hiển thị của giá trị (dùng khi có annot)
# cbar = True/False: hiển thị/không hiển thị thanh màu sắc
# vmin,vmax: giá trị nhỏ nhất, lớn nhất của thanh màu sắc
```

```python
sns.heatmap(data = fls, annot = True, fmt = 'd', cbar = True, vmin = 0, vmax = 700)
```
