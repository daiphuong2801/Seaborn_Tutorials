## 3.2. Trực quan hóa dữ liệu

### 3.2.1. Distribution Plot

### a. Histogram Plot

```python
#Cú pháp: sns.hisplot(data = Tên cần đặt_df, x = 'Chỉ tiêu').set(title = 'TÊN BIỂU ĐỒ')

#Các parameter bổ sung:
#     hue: phân biệt chỉ tiêu bất kỳ theo màu
#     hue_order: hoán đổi thứ tự của hue
#     multiple (layer, dodge, stack, fill): kiểu hiển thị của hue
#     stat (count, frequency, probability/proportion, percent, density): kiểu hiển thị giá trị ở trục tung (stat = 'count' => mặc định)
#     element (bars, step, poly): kiểu hiển thị của chỉ tiêu ở trục hoành 
#     kde = True/False: hiển thị/không hiển thị KDE Plot
```

```python
sns.histplot(data = tips_df, x = "total_bill", multiple = "stack",
             kde = False, element = 'bars', hue ='sex').set(title = 'TOTAL BILL COVERED BY GENDER')
```

### b. KDE (Kernel Density Estimate) Plot

```python
#Cú pháp: sns.kdeplot(data=Tên cần đặt_df, x = 'Chỉ tiêu').set(title = 'TÊN BIỂU ĐỒ')

#Các parameter bổ sung:
#     hue: phân biệt chỉ tiêu bất kỳ theo màu
#     hue_order: hoán đổi thứ tự của hue
#     multiple (layer, stack, fill): kiểu hiển thị của hue (multiple = 'layer' => mặc định)
```

```python
sns.kdeplot(data=tips_df, x ='total_bill', hue = 'sex', multiple = 'layer').set(title = 'TOTAL BILL COVERED BY GENDER')
```

### c. Displot

```python
#Cú pháp:
# 1 plot: 
#   sns.displot(data = Tên cần đặt_df, x = 'Chỉ tiêu').set(title = 'TÊN BIỂU ĐỒ')

# >= 2 plots (dùng col):
#   Tên tạm = sns.displot(data = Tên cần đặt_df, x = 'Chỉ tiêu')
#   Tên tạm.fig.suptitle('TÊN BIỂU ĐỒ') => Đặt tên biểu đồ
#   Tên tạm.fig.subplots_adjust(top/bottom = Kích thước) => Điều chỉnh kích thước tên biểu đồ

#Các parameter bổ sung:
#   col: phân biệt chỉ tiêu bất kỳ theo biểu đồ 
#   hue: phân biệt chỉ tiêu bất kỳ theo màu
#   hue_order: hoán đổi thứ tự của hue
#   multiple (layer, dodge, stack, fill): kiểu hiển thị giá trị ở trục tung 
#   kde = True/False: hiển thị/không hiển thị KDE Plot
#   kind (hist, kde, ecdf): kiểu biểu đồ muốn hiển thị (kind = 'hist' => mặc định)
#   stat (count, frequency, probability/proportion, percent, density): kiểu hiển thị giá trị ở trục tung (stat = 'count' => mặc định)
```

```python
tbill = sns.displot(data=tips_df, x = 'total_bill', col = 'time', 
                    hue = 'sex', multiple='stack', kde = True, kind = 'hist')

tbill.fig.suptitle('TOTAL BILL COVERED BY GENDER')
tbill.fig.subplots_adjust(top = .9)
```

### 3.2.5. FacetGrid

```python
#Cú pháp:
# 1 plot: 
#     Tên class = sns.FacetGrid(data = Tên cần đặt_df)
#     Tên class.map_dataframe(sns.histplot/scatterplot, x = 'Chỉ tiêu 1',
#                            y = 'Chỉ tiêu 2').set(title = 'TÊN BIỂU ĐỒ')

# >= 2 plots (dùng col):
#     Tên class = sns.FacetGrid(data = Tên cần đặt_df)
#     Tên class.map_dataframe(sns.histplot/scatterplot, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2')
#     Tên class.fig.suptitle('TÊN BIỂU ĐỒ') => Đặt tên biểu đồ
#     Tên class.fig.subplots_adjust(top/bottom = Kích thước) => Điều chỉnh kích thước tên biểu đồ

#Các parameter bổ sung:
#     col: phân biệt chỉ tiêu bất kỳ theo cột
#     row: phân biệt chỉ tiêu bất kỳ theo dòng
#     hue: phân biệt chỉ tiêu bất kỳ theo màu
#     hue/row/col_order: hoán đổi thứ tự của hue/row/col
```

```python
#Scatter Plot

fg = sns.FacetGrid(data=tips_df, hue = 'smoker')
fg.map_dataframe(sns.scatterplot, x = "total_bill", 
                 y = "tip").set(title = 'TIP AND TOTAL BILL COVERED BY SMOKER')
fg.add_legend() # hiển thị chú thích khi có hue
```

```python
#Histogram Plot

fg = sns.FacetGrid(data = tips_df, col = "time",  row = "sex")
fg.map_dataframe(sns.histplot, x = "total_bill")
fg.fig.suptitle('TOTAL BILL COVERED BY GENDER AND TIME')
fg.fig.subplots_adjust(top = .9)
```

### 3.2.6. Joint Plot

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
#Cú pháp: 
#     Tên tạm = sns.jointplot(data = Tên cần đặt_df, x = 'Chỉ tiêu 1', y = 'Chỉ tiêu 2')
#     Tên tạm.fig.suptitle('TÊN BIỂU ĐỒ')
#     Tên tạm.fig.subplots_adjust(top/bottom = kích thước)

#Các parameter bổ sung:
#     hue: phân biệt chỉ tiêu bất kỳ theo màu
#     hue_order: hoán đổi thứ tự của hue
#     kind (scatter, kde, hist, hex, reg, resid): kiểu biểu đồ muốn hiển thị
#                          (hue không đi kèm kind = 'hex', 'reg', 'resid')
#     xlim,ylim: khoảng giá trị được hiển thị trên trục x,y
```

```python
tb = sns.jointplot(data = tips_df, x = 'tip', y = 'total_bill', hue = 'smoker', kind = 'scatter', ylim = [0,60])
tb.fig.suptitle('CORRELATION BETWEEN TOTAL BILL AND TIP BY SMOKER') 
tb.fig.subplots_adjust(top = .9) 
```

### 3.2.7. Pair Plot

```python
pg_df = sns.load_dataset('penguins')
pg_df.tail()
```

```python
#Cú pháp:
#     Tên tạm = sns.pairplot(data = Tên cần đặt_df)
#     Tên tạm.fig.suptitle('TÊN BIỂU ĐỒ') => Đặt tên biểu đồ
#     Tên tạm.fig.subplots_adjust(top/bottom = kích thước) => Điều chỉnh kích thước tên biểu đồ

#Các parameter bổ sung:
#     hue: phân biệt chỉ tiêu bất kỳ theo màu
#     kind (scatter, kde, hist, reg): kiểu biểu đồ muốn hiển thị => mặc định: kind = 'scatter'
#     diag_kind (auto, hist, kde, None): kiểu biểu đồ muốn hiển thị đối với đường chéo chính
#     markers (o,s,D): thay đổi ký hiệu chú thích (của hue)
```

```python
pg = sns.pairplot(data = pg_df, hue = 'species', diag_kind = 'hist', markers = ['o','s','D'])
pg.fig.suptitle('CORRELATION BETWEEN BODY PARTS OF 3 PENGUINS') 
pg.fig.subplots_adjust(top = .9) 
```

### 3.2.8. Heat Map

```python
fl_df = sns.load_dataset('flights')
fl_df
```

```python
#Sắp xếp dữ liệu 

#Cú pháp: Tên table = pd.pivot_table(data = Tên cần đặt_df, index = 'Chỉ tiêu theo hàng',
#                     columns = 'Chỉ tiêu theo cột', values = 'Giá trị')
```

```python
fls = pd.pivot_table(data = fl_df, index = 'month', columns = 'year', values = 'passengers')
fls.head()
```

```python
#Cú pháp: sns.heatmap(data = Tên cần đặt) .set(title = 'TÊN BIỂU ĐỒ')

#Các parameter bổ sung:
# annot: hiển thị từng giá trị vào biểu đồ
# fmt (d,f): thay đổi kiểu hiển thị của giá trị (dùng khi có annot)
# cbar = True/False: hiển thị/không hiển thị thanh màu sắc
# vmin,vmax: giá trị nhỏ nhất, lớn nhất của thanh màu sắc
```

```python
sns.heatmap(data = fls, annot = True, fmt = 'd', cbar = True,
            vmin = 0, vmax = 700).set(title = 'PASSENGERS OF FLIGHTS FROM 1949 TO 1960')
```
